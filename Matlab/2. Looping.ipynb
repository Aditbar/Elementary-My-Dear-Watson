## While loop
#### akan melakukan pengulangan hingga kondisi true

```matlab
clear
clc
```

```matlab
% Vector increment
x = 0:5;
y = [];
i = 1; % butuh mendefinisikan i pada while agara nilai i nantinya bisa bertambah
```

```matlab
length(x)
```

```matlab
% mau looping i sebanyak jumlah data di array
while i<length(x)
    i
    i = i + 1
    disp("satu loop")
end
```

```matlab
clear
clc
```

```matlab
x = -5:.1:5;
y = [];
i = 1;
```

```matlab
while i<length(x)
    y(i) = x(i)
    i = i + 1;
end
```

```matlab
% jika sadar.......
% ada yang tidak beres disini sehingga
plot(x,y)
```

```matlab
% selalui ingat jumlah data antara x dan y harus sama untuk melakukan plot
% agar jumlah x dan y sama tambahkan <=
while i<=length(x)
    y(i) = x(i)
    i = i + 1;
end
```

```matlab
plot(x,y)
```

## break
ketika break akan keluar dari looping

```matlab
for i=1:20
    if i ==10
        disp(i);
        disp("ketemu angka 10")
        break
    end
    disp(i)
    disp("Budi lewat")
end
disp("akhir looping");

% angka 11 sampai 20 tidak di print
```

## Continue

```matlab
for i=1:20
    if i ==10
        disp(i);
        disp("ketemu angka 10")
        continue
    end
    disp(i)
    disp("Amir lewat")
end
disp("akhir looping");

% angka 11 sampai 20 tidak di print
```
