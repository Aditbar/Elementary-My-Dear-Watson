```matlab
% membuat 2 buat grafik
clear
clc

x = linspace(0,2*pi,100);
y1 = sin(x);
y2 = cos(x);

figure(1)
plot(x,y1)

figure(2)
plot(x,y1,x,y2)
```

```matlab
% warna custom
clear 
clc

x = linspace(0,2*pi,100);
y1 = sin(x);
y2 = cos(x);
y3 = sin(x+1);

plot(x,y1,'b',x,y2,'r',x,y3,'g')
```

```matlab
% warna dan tipe garis custom
clear 
clc

x = linspace(0,2*pi,100);
y1 = sin(x);
y2 = cos(x);
y3 = sin(x+1);

plot(x,y1,'b--',x,y2,'r-.',x,y3,'m:')
```

```matlab
% labelin
clear 
clc

x = linspace(0,2*pi,100);
y1 = sin(x);
plot(x,y1)
axis([0 2*pi -1 1]);
title("plot sin(x)");
xlabel("sudut")
ylabel("amplitudo")
```
